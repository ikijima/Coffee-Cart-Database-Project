I have a database of a coffee cart on the go business, with a total of 10 carts in operation in Central Jakarta and South Jakarta, with each cart carries about 200 cups of ready-to-serve coffee. The menu consists of Iced Americano, Iced Latte, Iced Vanilla Latte, Iced Hazelnut Latte, and Iced Aren Latte (aren is a palm sugar made into syrup).

Provided here is the database structure.
```SQL
CREATE TABLE carts (
cart_id serial PRIMARY KEY,
cart_name varchar(50),
LOCATION varchar(100),
status varchar(20) -- Active, Inactive
);

CREATE TABLE products (
product_id serial PRIMARY KEY,
product_name varchar(50),
price decimal(6,2),
description TEXT,
stock_level int
);

CREATE TABLE orders (
    order_id UUID PRIMARY KEY,
    customer_id UUID,  -- Optional if you want to track customers
    sale_timestamp TIMESTAMP,
    payment_method VARCHAR(20),  -- e.g., 'Cash', 'Card', 'E-wallet'
    cart_id INT REFERENCES carts(cart_id),
    location VARCHAR(100)
);

CREATE TABLE order_items (
    order_item_id UUID PRIMARY KEY,
    order_id UUID REFERENCES orders(order_id),
    product_id INT REFERENCES products(product_id),
    quantity INT DEFAULT 1,  -- Quantity of the product sold
    price DECIMAL(10, 2)
);

CREATE TABLE sales (
sale_id serial PRIMARY KEY,
cart_id INT REFERENCES carts(cart_id),
product_id INT REFERENCES products(product_id),
quantity int,
total_price decimal(10,2),
sale_timestamp timestamp DEFAULT current_timestamp,
payment_method varchar(20) -- cash, gopay, ovo, qris bank
);

CREATE TABLE inventory (
inventory_id serial PRIMARY KEY,
cart_id int REFERENCES carts(cart_id),
product_id int REFERENCES products(product_id),
stock_level int,
last_restock timestamp
);

CREATE TABLE employees (
employee_id serial PRIMARY KEY,
employee_name varchar(100),
emp_call_name varchar(20),
cart_id int REFERENCES carts(cart_id),
ROLE varchar(20)
);

CREATE TABLE raw_materials (
raw_mat_id serial PRIMARY KEY,
material_name varchar(50), -- coffee beans, uht, milk, etc
unit varchar(20), -- kg, liter, etc
stock_level decimal(10,2) -- amount OF material IN stock
reorder_threshold decimal(10,2) -- LEVEL OF material TO be restocked
);

CREATE TABLE suppliers (
    supplier_id SERIAL PRIMARY KEY,
    supplier_name VARCHAR(100),
    contact_info TEXT
);

CREATE TABLE raw_materials_purchases (
purchase_id serial PRIMARY KEY,
raw_mat_id int REFERENCES raw_materials(raw_mat_id),
supplier_id int REFERENCES suppliers(supplier_id),
quantity decimal(10,2),
unit_price decimal(10,2),
purchase_date date,
total_cost AS (quantity * unit_price),
received_date date
);

CREATE TABLE raw_material_consumption (
    consumption_id SERIAL PRIMARY KEY,
    raw_material_id INT REFERENCES raw_materials(raw_material_id),
    cart_id INT REFERENCES carts(cart_id), -- Optional: Track consumption per cart
    quantity_used DECIMAL(10,2),
    consumption_date DATE
);

CREATE TABLE recipes (
    recipe_id SERIAL PRIMARY KEY,
    product_id INT REFERENCES products(product_id),
    raw_material_id INT REFERENCES raw_materials(raw_material_id),
    quantity_per_unit DECIMAL(10,2) -- Amount of raw material used per product unit
);
```

Below are Python scripts using the `Faker` library to generate CSV files for the tables related to sales: `orders`, `order_items`, `sales`, `inventory`, `raw_materials_purchases`, and `raw_material_consumption`. You'll need to have the `Faker` library installed, which you can do with `pip install faker`.

### Python Script to Generate CSV Files

```python
import csv
import random
from faker import Faker
import uuid
from datetime import datetime, timedelta

fake = Faker()

# Constants
num_orders = 50
num_order_items = 100
num_sales = 50
num_inventory_records = 10
num_raw_material_purchases = 20
num_raw_material_consumption = 30
num_carts = 10
num_products = 5
num_raw_materials = 5
num_suppliers = 3

# Generate Orders
def generate_orders():
    orders = []
    for _ in range(num_orders):
        order_id = uuid.uuid4()
        customer_id = uuid.uuid4()
        sale_timestamp = fake.date_time_between(start_date='-30d', end_date='now')
        payment_method = random.choice(['Cash', 'Card', 'E-wallet'])
        cart_id = random.randint(1, num_carts)
        location = fake.city()

        orders.append([str(order_id), str(customer_id), sale_timestamp, payment_method, cart_id, location])

    with open('orders.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['order_id', 'customer_id', 'sale_timestamp', 'payment_method', 'cart_id', 'location'])
        writer.writerows(orders)

# Generate Order Items
def generate_order_items(orders):
    order_items = []
    for _ in range(num_order_items):
        order_id = random.choice(orders)[0]
        product_id = random.randint(1, num_products)
        quantity = random.randint(1, 5)
        price = round(quantity * random.uniform(25, 35), 2)  # assuming products are in that range

        order_items.append([str(uuid.uuid4()), order_id, product_id, quantity, price])

    with open('order_items.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['order_item_id', 'order_id', 'product_id', 'quantity', 'price'])
        writer.writerows(order_items)

# Generate Sales
def generate_sales(orders):
    sales = []
    for _ in range(num_sales):
        cart_id = random.randint(1, num_carts)
        product_id = random.randint(1, num_products)
        quantity = random.randint(1, 5)
        total_price = round(quantity * random.uniform(25, 35), 2)  # assuming products are in that range
        sale_timestamp = fake.date_time_between(start_date='-30d', end_date='now')
        payment_method = random.choice(['Cash', 'Gopay', 'OVO', 'QRIS', 'Bank'])

        sales.append([len(sales) + 1, cart_id, product_id, quantity, total_price, sale_timestamp, payment_method])

    with open('sales.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['sale_id', 'cart_id', 'product_id', 'quantity', 'total_price', 'sale_timestamp', 'payment_method'])
        writer.writerows(sales)

# Generate Inventory
def generate_inventory():
    inventory_records = []
    for cart_id in range(1, num_carts + 1):
        for product_id in range(1, num_products + 1):
            stock_level = random.randint(100, 200)  # initial stock levels
            last_restock = fake.date_time_this_year()

            inventory_records.append([len(inventory_records) + 1, cart_id, product_id, stock_level, last_restock])

    with open('inventory.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['inventory_id', 'cart_id', 'product_id', 'stock_level', 'last_restock'])
        writer.writerows(inventory_records)

# Generate Raw Material Purchases
def generate_raw_material_purchases():
    purchases = []
    for _ in range(num_raw_material_purchases):
        raw_mat_id = random.randint(1, num_raw_materials)
        supplier_id = random.randint(1, num_suppliers)
        quantity = round(random.uniform(1, 50), 2)
        unit_price = round(random.uniform(5, 20), 2)
        purchase_date = fake.date_between(start_date='-30d', end_date='now')
        total_cost = quantity * unit_price
        received_date = purchase_date + timedelta(days=random.randint(1, 3))

        purchases.append([len(purchases) + 1, raw_mat_id, supplier_id, quantity, unit_price, purchase_date, total_cost, received_date])

    with open('raw_materials_purchases.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['purchase_id', 'raw_mat_id', 'supplier_id', 'quantity', 'unit_price', 'purchase_date', 'total_cost', 'received_date'])
        writer.writerows(purchases)

# Generate Raw Material Consumption
def generate_raw_material_consumption():
    consumption_records = []
    for _ in range(num_raw_material_consumption):
        raw_material_id = random.randint(1, num_raw_materials)
        cart_id = random.randint(1, num_carts)
        quantity_used = round(random.uniform(0.1, 5), 2)
        consumption_date = fake.date_between(start_date='-30d', end_date='now')

        consumption_records.append([len(consumption_records) + 1, raw_material_id, cart_id, quantity_used, consumption_date])

    with open('raw_material_consumption.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['consumption_id', 'raw_material_id', 'cart_id', 'quantity_used', 'consumption_date'])
        writer.writerows(consumption_records)

# Generate all CSVs
generate_orders()
generate_order_items(orders)
generate_sales(orders)
generate_inventory()
generate_raw_material_purchases()
generate_raw_material_consumption()
```

4. **File Creation**: The CSV files are created in the same directory as the script.
